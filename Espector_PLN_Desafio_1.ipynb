# 2. Construir un modelo de clasificación por prototipos (tipo zero-shot).
Clasificar los documentos de un conjunto de test comparando cada uno con todos los de entrenamiento y asignar la clase al label del documento del conjunto de entrenamiento con mayor similaridad.

```python
X_test = tfidfvect.transform(newsgroups_test.data)
y_test = newsgroups_test.target
```

```python
y_pred = []
for x in X_test:
  cossim = cosine_similarity(x, X_train)[0]
  mostsim = np.argmax(cossim)
  y_pred.append(y_train[mostsim])
```

```python
f1_zeroshot=f1_score(y_test, y_pred, average='macro')
print("El f1-score del modelo de clasificación tipo zero-shot es",f1_zeroshot)
```

# 4. Transponer la matriz documento-término.
De esa manera se obtiene una matriz término-documento que puede ser interpretada como una colección de vectorización de palabras. Estudiar ahora similaridad entre palabras tomando 5 palabras y estudiando sus 5 más similares. **La elección de palabras no debe ser al azar para evitar la aparición de términos poco interpretables, elegirlas "manualmente".**

Para este ejercicio, se vuelve a vectorizar utilizando los valores por defecto y se transpone la matriz documento-término:

```python
tfidfvect = TfidfVectorizer()
X_train = tfidfvect.fit_transform(newsgroups_train.data)
y_train = newsgroups_train.target
X_test = tfidfvect.transform(newsgroups_test.data)
y_test = newsgroups_test.target
```

```python
X_train_T = X_train.T
```

Se obtiene diccionario que va de índices a términos:

```python
idx2word = {v: k for k,v in tfidfvect.vocabulary_.items()}
```

Se seleccionan palabras que fueron observadas como clave en los textos analizados previamente:

```python
palabras_elegidas = ['weapons', 'motorcycle', 'program', 'captain', 'nuclear']
```

```python
indices_elegidos = [tfidfvect.vocabulary_[word] for word in palabras_elegidas]
```

Se calcula la similaridad entre las palabras elegidas y el resto de las palabras utilizando la similaridad coseno:

```python
word_similarity = cosine_similarity(X_train_T[indices_elegidos], X_train_T)
```

Obtener las 5 palabras más similares a cada una de las palabras elegidas:

```python
for i, word_index in enumerate(indices_elegidos):
    similarities = word_similarity[i]

    most_similar_indices = np.argsort(similarities)[::-1][1:6]

    print(f"Palabras más similares a '{palabras_elegidas[i]}':")
    for idx in most_similar_indices:
        print(f"- {idx2word[idx]}")
    print("-" * 40)
```

En general, se observa que las palabras obtenidas como más similares en cada caso tienen sentido. En algunos casos, como 'weapons', 'program' y 'nuclear' las relaciones con las palabras están claras. En otros casos, la relación se ve más claramente si analizamos los textos que se estudiaron previamente.

Por ejemplo, en el caso de 'motorcycle', la palabra 'inquiries' en principio no tendría sentido, pero los textos analizados correspondían a preguntas/consultas que hacían los usuarios sobre las motos, por lo que allí podría estar la relación.

Otro ejemplo es el caso de 'captain', en donde vemos algunos nombres propios ('Tochett', 'Chorske'). Analizando los textos previos, vemos que esos nombres corresponden a capitanes de hockey. Si en los datos hay muchos documentos sobre hockey, entonces tiene sentido que los términos estén relacionados.
